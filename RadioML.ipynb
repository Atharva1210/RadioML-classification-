```python
import os
import zipfile
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.signal
import tensorflow as tf

from tensorflow.keras import layers, models, utils
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, f1_score, precision_score, recall_score

zip_path = "/content/archive (5).zip"
extracted_folder = "./RML2016.10a_dataset"

if not os.path.exists(extracted_folder):
    print("Extracting dataset from zip file...")
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extracted_folder)
    print("Extraction complete!")
else:
    print("Dataset already extracted.")

pkl_file = os.path.join(extracted_folder, "RML2016.10a_dict.pkl")
print("Loading dataset from:", pkl_file)

with open(pkl_file, 'rb') as f:
    data = pickle.load(f, encoding='latin1')


X_list = []
Y_list = []
for (mod, snr) in sorted(data.keys()):
    if snr >= 0:
        samples = data[(mod, snr)]
        X_list.append(samples)
        Y_list.extend([mod] * samples.shape[0])

X = np.concatenate(X_list)
Y = np.array(Y_list)
print("Total samples after filtering SNR>=0:", X.shape[0])


X = np.transpose(X, (0, 2, 1))
print("Sample shape after transpose:", X.shape[1:])

desired_count = 100_000
if X.shape[0] > desired_count:
    print(f"Downsampling from {X.shape[0]} to {desired_count} samples...")
    idx = np.random.choice(X.shape[0], desired_count, replace=False)
    X = X[idx]
    Y = Y[idx]
print("Final sample count:", X.shape[0])

mod_list = np.unique(Y)
num_classes = len(mod_list)
print("Modulation types:", mod_list)
mod_to_idx = {m: i for i, m in enumerate(mod_list)}
Y_int = np.array([mod_to_idx[m] for m in Y])

def iq_to_spectrogram(iq_sample, nperseg=32, noverlap=16, nfft=32):
    """
    Convert a single IQ sample (shape [128,2]) to a spectrogram.
    Combines I and Q into a complex signal.
    Computes the STFT and takes only the positive frequencies.
    """

    complex_signal = iq_sample[:, 0] + 1j * iq_sample[:, 1]
    freqs, times, Zxx = scipy.signal.stft(
        complex_signal,
        window='hann',
        nperseg=nperseg,
        noverlap=noverlap,
        nfft=nfft,
        boundary=None
    )
    Sxx = np.abs(Zxx)

    Sxx = Sxx[: nfft//2 + 1, :]
    return Sxx

print("Generating spectrograms...")
spectrograms = []
for i in range(X.shape[0]):
    spec = iq_to_spectrogram(X[i], nperseg=32, noverlap=16, nfft=32)
    spectrograms.append(spec)
spectrograms = np.array(spectrograms)
print("Spectrogram shape:", spectrograms.shape)

spectrograms = spectrograms[..., np.newaxis]
print("Spectrogram shape with channel:", spectrograms.shape)

Y_cat = utils.to_categorical(Y_int, num_classes=num_classes)
X_train, X_test, Y_train, Y_test, Y_train_int, Y_test_int = train_test_split(
    spectrograms, Y_cat, Y_int, test_size=0.2, random_state=42, stratify=Y_int
)
print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])


input_shape_2d = X_train.shape[1:]
print("Input shape for CNN:", input_shape_2d)

model = models.Sequential([
    tf.keras.Input(shape=input_shape_2d),

    # First Conv Block
    layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(pool_size=(2, 1)),
    layers.Dropout(0.3),

    # Second Conv Block
    layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(pool_size=(2, 1)),
    layers.Dropout(0.3),

    # Third Conv Block
    layers.Conv2D(256, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.Conv2D(256, (3, 3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D(pool_size=(2, 1)),
    layers.Dropout(0.4),

    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(num_classes, activation='softmax')
])
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

epochs = 40
batch_size = 128

early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(
    X_train, Y_train,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.1,
    callbacks=[early_stop],
    verbose=1
)


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], 'b-o', label='Train Accuracy')
plt.plot(history.history['val_accuracy'], 'r-o', label='Val Accuracy')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], 'b-o', label='Train Loss')
plt.plot(history.history['val_loss'], 'r-o', label='Val Loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()

test_loss, test_acc = model.evaluate(X_test, Y_test, verbose=0)
print(f"Test Accuracy: {test_acc*100:.2f}%")

Y_pred = model.predict(X_test)
Y_pred_int = np.argmax(Y_pred, axis=1)

print("\nClassification Report:")
print(classification_report(Y_test_int, Y_pred_int, target_names=mod_list))
print(f"Weighted F1 Score: {f1_score(Y_test_int, Y_pred_int, average='weighted'):.4f}")
print(f"Weighted Precision: {precision_score(Y_test_int, Y_pred_int, average='weighted'):.4f}")
print(f"Weighted Recall: {recall_score(Y_test_int, Y_pred_int, average='weighted'):.4f}")

cm = confusion_matrix(Y_test_int, Y_pred_int)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=mod_list, yticklabels=mod_list)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()


model_save_path = "modulation_classifier_highSNR_90plus.h5"
model.save(model_save_path)
print("Model saved as:", model_save_path)
```

```python
import os
import numpy as np
import pickle
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Path to your dataset
zip_path = '/content/archive (5).zip'
extracted_folder = './RML2016.10a_dataset'

# Extract the dataset if not already extracted
if not os.path.exists(extracted_folder):
    print("Extracting dataset from zip file...")
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extracted_folder)
    print("Extraction complete!")
else:
    print("Dataset already extracted.")

# Load the RadioML dataset
pkl_file = os.path.join(extracted_folder, "RML2016.10a_dict.pkl")
print("Loading dataset from:", pkl_file)

with open(pkl_file, 'rb') as f:
    data = pickle.load(f, encoding='latin1')

# Collect samples (X) and labels (Y)
X_list = []
Y_list = []
for (mod, snr) in sorted(data.keys()):
    samples = data[(mod, snr)]  # shape: (num_samples, 2, 128)
    X_list.append(samples)
    Y_list += [mod] * samples.shape[0]

X = np.concatenate(X_list)  # shape => (total_samples, 2, 128)
Y = np.array(Y_list)

# Preprocess: Transpose from (2,128) to (128, 2) and normalize
X = np.transpose(X, (0, 2, 1))  # shape => (num_samples, 128, 2)

# Use LabelEncoder to convert modulation names into integer labels
label_encoder = LabelEncoder()
Y_int = label_encoder.fit_transform(Y)

# Folder where sample images will be saved
output_dir = "sample_images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Assuming:
# - spectrograms is a NumPy array of shape (num_samples, height, width, 1)
# - Y is a NumPy array of shape (num_samples,) containing modulation names (e.g., 'BPSK', 'QPSK', etc.)
# - mod_list is an array/list of unique modulation names
mod_list = np.unique(Y)

# For each modulation class, extract 10 random images
for mod in mod_list:
    # Find indices for this class
    indices = np.where(Y == mod)[0]
    # If there are fewer than 10 samples, take all; otherwise, randomly select 10
    num_to_extract = 10 if len(indices) >= 10 else len(indices)
    selected_indices = np.random.choice(indices, num_to_extract, replace=False)

    for count, idx in enumerate(selected_indices):
        # Get the spectrogram image (no need to squeeze anymore)
        img = X[idx]  # Shape: (128, 2)

        # Plot the spectrogram and save it as an image
        plt.figure(figsize=(4,3))
        plt.imshow(img, aspect='auto', origin='lower', cmap='viridis')
        plt.title(f"{mod} sample {count+1}")
        plt.colorbar()

        # Save the figure
        filename = os.path.join(output_dir, f"{mod}_sample_{count+1}.png")
        plt.savefig(filename)
        plt.close()
        print(f"Saved: {filename}")
```

```python


# Import necessary libraries
import numpy as np
import tensorflow as tf
from PIL import Image
import librosa
import librosa.display
import matplotlib.pyplot as plt
from google.colab import files

# Load your trained model (adjusted to your provided path)
model_path = '/content/modulation_classifier_highSNR_90plus.h5'  # Model path in Colab
model = tf.keras.models.load_model(model_path)

# Print the model summary to inspect the input shape
model.summary()

# Define the modulation class labels (update with your model's labels if needed)
CLASS_LABELS = ["8PSK", "AM-DSB", "AM-SSB", "BPSK", "CPFSK", "GFSK", "PAM4", "QAM16", "QAM64", "QPSK", "WBFM"]

# Function to upload and classify spectrogram images
def classify_and_reconstruct():
    # Upload the spectrogram image
    uploaded_image = files.upload()
    image_path = list(uploaded_image.keys())[0]  # Get the file name

    # Load the uploaded image
    image = Image.open(image_path).convert("L")  # Convert to grayscale
    plt.imshow(image, cmap='viridis')
    plt.title("Uploaded Spectrogram")
    plt.colorbar()
    plt.show()

    # Preprocess the image for classification:
    # Assuming the model expects a spectrogram with shape (7, 17)
    target_size = (7, 17)  # (width, height)
    image_resized = image.resize(target_size)
    image_array = np.array(image_resized).astype("float32")  # No normalization yet

    # Try Normalizing to [0, 1]
    image_array /= 255.0  # Normalize to [0, 1]
    # If model was trained with [-1, 1], try this:
    # image_array = (image_array / 127.5) - 1  # Normalize to [-1, 1]

    # For the CNN, we need shape (1, height, width, 1)
    input_img = np.expand_dims(image_array, axis=(0, -1))

    # Classify using the model
    preds = model.predict(input_img)
    pred_index = np.argmax(preds, axis=1)[0]
    pred_label = CLASS_LABELS[pred_index]
    confidence = preds[0][pred_index]

    # Output the prediction result
    print(f"Predicted Modulation: {pred_label} (Confidence: {confidence*100:.2f}%)")

    # For signal reconstruction, we use the resized image as the magnitude spectrogram.
    magnitude = image_array  # shape: (7, 17)

    # Griffin-Lim reconstruction:
    # Set n_fft such that n_fft//2+1 equals the number of frequency bins (here, 17)
    n_fft = (17 - 1) * 2  # 32; since 32//2+1 = 17
    hop_length = 16        # This choice gives roughly 7 time frames for an original 128-length signal

    # Griffin-Lim expects the magnitude spectrogram; it will estimate a phase.
    reconstructed_signal = librosa.griffinlim(magnitude, n_fft=n_fft, hop_length=hop_length)

    # Plot the reconstructed signal waveform
    fig, ax = plt.subplots(figsize=(8, 4))
    sr = 16000  # Assume a sample rate (for display purposes) of 16kHz
    librosa.display.waveshow(reconstructed_signal, sr=sr, ax=ax)
    ax.set_title("Reconstructed Signal Waveform")
    plt.show()

# Call the function to upload the image and classify it
classify_and_reconstruct()
```
